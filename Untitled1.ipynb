```python
# =====================================================
# AI STOCK MARKET PREDICTION SYSTEM (LSTM DEEP LEARNING)
# =====================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
import math

np.random.seed(42)

# -----------------------------------------
# 1. GENERATE SYNTHETIC STOCK DATA
# -----------------------------------------

def generate_stock_data(days=1000):
    price = 100
    prices = []
    for _ in range(days):
        daily_return = np.random.normal(0.001, 0.02)
        price = price * (1 + daily_return)
        prices.append(price)
    return prices

stock_prices = generate_stock_data()

df = pd.DataFrame(stock_prices, columns=["Close"])
df["Day"] = range(len(df))

print("Dataset Preview:")
print(df.head())

# -----------------------------------------
# 2. VISUALIZE STOCK TREND
# -----------------------------------------

plt.figure(figsize=(12,6))
plt.plot(df["Close"])
plt.title("Synthetic Stock Price Trend")
plt.xlabel("Days")
plt.ylabel("Price")
plt.show()

# -----------------------------------------
# 3. DATA PREPROCESSING
# -----------------------------------------

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(df["Close"].values.reshape(-1,1))

def create_sequences(data, seq_length=60):
    X = []
    y = []
    for i in range(seq_length, len(data)):
        X.append(data[i-seq_length:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

X, y = create_sequences(scaled_data)

X = np.reshape(X, (X.shape[0], X.shape[1], 1))

# -----------------------------------------
# 4. TRAIN TEST SPLIT
# -----------------------------------------

split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# -----------------------------------------
# 5. BUILD LSTM MODEL
# -----------------------------------------

model = Sequential()

model.add(LSTM(units=64, return_sequences=True, input_shape=(X_train.shape[1],1)))
model.add(Dropout(0.2))

model.add(LSTM(units=64, return_sequences=False))
model.add(Dropout(0.2))

model.add(Dense(units=25))
model.add(Dense(units=1))

model.compile(optimizer="adam", loss="mean_squared_error")

print(model.summary())

# -----------------------------------------
# 6. TRAIN MODEL
# -----------------------------------------

history = model.fit(X_train, y_train, epochs=5, batch_size=32)

# -----------------------------------------
# 7. MODEL EVALUATION
# -----------------------------------------

predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions.reshape(-1,1))
y_test_actual = scaler.inverse_transform(y_test.reshape(-1,1))

mse = mean_squared_error(y_test_actual, predictions)
rmse = math.sqrt(mse)

print("\nModel RMSE:", rmse)

# -----------------------------------------
# 8. VISUALIZE PREDICTIONS
# -----------------------------------------

plt.figure(figsize=(12,6))
plt.plot(y_test_actual, label="Actual Price")
plt.plot(predictions, label="Predicted Price")
plt.legend()
plt.title("Stock Price Prediction vs Actual")
plt.show()

# -----------------------------------------
# 9. FUTURE FORECASTING
# -----------------------------------------

last_60_days = scaled_data[-60:]
future_input = last_60_days.reshape(1,60,1)

future_prices = []

for _ in range(7):  # Predict next 7 days
    next_price = model.predict(future_input)
    future_prices.append(next_price[0][0])
    # Correct the reshaping of next_price to match the dimensions for concatenation
    future_input = np.append(future_input[:,1:,:], next_price.reshape(1,1,1), axis=1)

future_prices = scaler.inverse_transform(np.array(future_prices).reshape(-1,1))

print("\nNext 7 Days Predicted Prices:")
for i, price in enumerate(future_prices):
    print(f"Day {i+1}: {price[0]:.2f}")

# -----------------------------------------
# 10. VOLATILITY ANALYSIS
# -----------------------------------------

returns = df["Close"].pct_change()
volatility = returns.std() * np.sqrt(252)

print("\nEstimated Annual Volatility:", volatility)

plt.figure(figsize=(10,5))
sns.histplot(returns.dropna(), bins=50, kde=True)
plt.title("Return Distribution")
plt.show()

# -----------------------------------------
# 11. INVESTMENT SIMULATION
# -----------------------------------------

initial_investment = float(input("\nEnter investment amount: "))
shares = initial_investment / df["Close"].iloc[-1]
future_value = shares * future_prices[-1][0]

print("\nEstimated Portfolio Value After 7 Days:", round(future_value,2))
```
